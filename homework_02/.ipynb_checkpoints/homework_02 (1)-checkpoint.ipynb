```python
for n in range(99,0,-1):
        if n == 0:
            print ("No more bottles of beer on the wall, no more bottles of beer.")
            print ("Go to the store and buy some more,99 bottles of beer on the wall")
        elif n == 1:
            print ("1 bottle of beer on the wall, 1 bottle of beer.")
            print ("Take one down and pass it around, no more bottles of beer ")
        elif n == 2:
            print (n, " bottles of beer on the wall, " ,n, " bottles of beer.")
            print ("Take one down and pass it around, " ,n-1, " bottle of beer on the wall.")
        else :
            print (n, " bottles of beer on the wall, " ,n, " bottles of beer.")
            print ("Take one down and pass it around, " ,n-1, " bottles of beer on the wall.")
```

```python

for n in range(1, 101):
    if n % 3 == 0 and n % 5 == 0:
        print("FizzBuzz")
    elif n % 3 == 0:
        print("Fizz")
    elif n % 5 == 0:
        print("Buzz")
    else:
        print(n)
 
```
